```python
import pandas as pd
```

```python
df = pd.read_csv('./clipper_sorted_racetags.csv').drop(['Unnamed: 0'], axis=1)
```

```python
df_clean = df.drop_duplicates()
```

```python
df_clean.reset_index().drop(['index'], axis=1)
```

```python
df_clean.to_csv('./clipper-clean.csv')
```

```python
import os
os.mkdir('./races')
```

```python
for race in df_clean['race'].unique():
    df_clean[df_clean['race'] == race].round(4).to_csv('./races/clipper-' + race + '.csv')
```
