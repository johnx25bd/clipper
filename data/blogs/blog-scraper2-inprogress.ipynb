```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
url = 'https://www.clipperroundtheworld.com/skipper-report/visit-seattle/race1-day8-team26'

response = requests.get(url)
content = response.content
```

```python
soup = bs(content)
```

```python
url.split('/')
```

```python
output = []


url = url.split('/')
post = {}
post['team'] = url[-2]
post['writer'] = url[-3].split('-')[0]
post['race'] = url[-1].split('-')[0].replace('race', '')
post['day'] = url[-1].split('-')[1].replace('day', '')
post['text'] = []

for p in soup.find_all('div', class_='article')[0].find_all('p'):
    post['text'].append(p.text)
    
print(post)
```
