```python
import pandas as pd
import geojson
import pprint
import json
```

```python
df = pd.read_csv('./clipper-time-indexed.csv').drop(columns=['Unnamed: 0'], axis=1)
```

```python
east_na = df[df['great-britain_race'] == 'east-north-america']
```

```python
east_na = east_na.reset_index().drop(columns=['index'], axis=1).drop_duplicates().dropna()
```

```python
teams = [team.split('_')[0] for team in east_na.columns if '-id' in team]
```

```python
# Getting list of fleet centroids
lat_cols = [col for col in east_na.columns if 'lat' in col]
lon_cols = [col for col in east_na.columns if 'lon' in col]

east_na['lon_centroid'] = east_na[lon_cols].mean(axis=1)
east_na['lat_centroid'] = east_na[lat_cols].mean(axis=1)
```

```python
east_na[east_na['unicef_team'] != "Unicef"][[col for col in east_na.columns if 'icef' in col]]
```

```python
fleet_centroids = []
for row in east_na.iterrows():
    print( round(row[1]['sanya-serenity-coast_lat'], 5))
    fleet_centroids.append([round(row[1]['lon_centroid'], 5), round(row[1]['lat_centroid'], 5)])
```

```python
timestamps = east_na['sanya-serenity-coast_timestamp'].tolist()
```

```python
for team in teams:
    cols = [col for col in east_na.columns if team in col]
    
```

```python
features = []
for team_id_string in teams:
    ship = east_na[['race', 'datetime', 'sanya-serenity-coast_timestamp'] + [col for col in east_na.columns if team_id_string in col]]
    ship = ship.fillna(-999)

    ship_positions = list(ship[team_id_string + '_position'])
    team = ship[team_id_string + '_team'].unique()[0]
    dtf = list(ship[team_id_string + '_dtf'])
    dist_mg = list(ship[team_id_string + '_dist_mg'])

    ship_properties = {
        "team": team, 
        "team-id": team_id_string, 
        "positions": ship_positions, 
        "dtf": dtf,
        "dist_mg": dist_mg,
        "timestamps": timestamps
    }
    
    print(team, team_id_string)

    line = []
    for row in ship[[team_id_string + '_lon', team_id_string + '_lat']].iterrows():
        lon = row[1][0].round(5)
        lat = row[1][1].round(5)
        line.append([lon,lat])

    ship_feature = geojson.Feature(geometry=geojson.LineString(line), properties=ship_properties)
    features.append(ship_feature)
```

```python
pprint.pprint(features)
```

```python
collection = geojson.FeatureCollection(features)
```

```python
metadata = {
    "raceID": "eastNA", 
    "raceName": "Panama to New York: Eastern North America", 
    "fleetCentroids": fleet_centroids,
    "finish": [-73.952224, 40.511710]
}
collection.metadata = metadata
```

```python
with open('eastNA-metadata.json', 'w') as output:
    json.dump(collection, output)
```
